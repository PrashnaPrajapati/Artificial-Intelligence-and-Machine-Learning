Create a Python program that converts between different units of measurement.

– Convert pounds (lbs) to kilograms (kg).

– Convert liters (L) to gallons (gal).

– Convert gallons (gal) to liters (L)

```python
def unit_convertor():
  print("Choose conversion type:")
  print("1. Length")
  print("2. Weight")
  print("3. Volume")

  choice = input("Enter 1, 2, or 3: ")

  try:
    value = float(input("Enter value to convert:"))

    if choice == "1":
      print("1. Meters to Feet")
      print("2. Feet to Meters")
      sub = input("Choose option: ")

      if sub == "1":
        result = value * 3.28084
        print("Result:", result, "feet")
        return result

      elif sub == "2":
        result = value / 3.28084
        print("Result:", result, "meters")
        return result

      else:
        print("Invalid length option.")
        return None

    elif choice == "2":
      print("1. Kg to Pounds")
      print("2. Pounds to Kg")
      sub = input("Choose option: ")

      if sub == "1":
          result = value * 2.20462
          print("Result:", result, "lbs")
          return result

      elif sub == "2":
          result = value / 2.20462
          print("Result:", result, "kg")
          return result

      else:
          print("Invalid weight option.")
          return None

    elif choice == "3":
      print("1. Liters to Gallons")
      print("2. Gallons to Liters")
      sub = input("Choose option:")

      if sub == "1":
        result = value * 0.264172
        print("Result:", result, "gallons")
        return result

      elif sub == "2":
        result = value / 0.264172
        print("Result:", result, "liters")
        return result

      else:
        print("Invalid volume option")
        return None

    else:
      print("Invalid main choice.")
      return None

  except ValueError:
    print("Invalid input. Please enter numbers only.")
    return None

unit_convertor()
```

#5. Find the Maximum Element in a Nested List:

Write a Python function that finds the maximum element in a nested list (regardless of depth).

• Requirements:

– Define a function find_max(lst) that takes a nested list lst and returns the maximum elememt.

– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    max_val = float('-inf')
    for item in lst:
        if isinstance(item, type([])):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)
    return max_val

nested_list = [[1, 2], [3, [4, 5]], 6]
max_element = find_max(nested_list)

print("Input nested list:", nested_list)
print("Maximum element:", max_element)
```

#7. Flatten a List of Lists of Lists:

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

• Requirements:

– Define a function deep_flatten(lst) that takes a deeply nested list lst and returns a single flattened list.

– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be [1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
  flat_list = []
  for item in lst:
    if isinstance(item, type([])):
      flat_list.extend(deep_flatten(item))
    else:
      flat_list.append(item)
  return flat_list

nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)

print("Input nested list:", nested_list)
print("Flattened list:", result)
```

#8. Nested List Average:

Write a Python function that calculates the average of all elements in a nested list.

• Requirements:

– Define a function average_nested(lst) that takes a nested list lst and returns the average of all the elements.

– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
    def sum_and_count(sublist):
        total_sum = 0
        total_count = 0
        for item in sublist:
            if isinstance(item, type([])):
                s, c = sum_and_count(item)
                total_sum += s
                total_count += c
            else:
                total_sum += item
                total_count += 1
        return total_sum, total_count

    total_sum, total_count = sum_and_count(lst)
    return total_sum / total_count if total_count > 0 else 0

nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)

print("Input nested list:", nested_list)
print("Average of all elements:", result)
```

#3. Dot Product

• Using Python Lists, compute the dot product of two lists of size 1,000,000. Measure and
Print the time taken for this operation.


```python
import time
import numpy as np

list1 = range(1000000)
list2 = range(1000000)

start = time.time()

dot_product_list = sum(a * b for a, b in zip(list1, list2))

end = time.time()

print("Dot product using Python lists:", dot_product_list)
print("Time taken using Python lists:", end - start, "seconds")
```

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for

```python
array1 = np.arange(1000000)
array2 = np.arange(1000000)

start = time.time()

dot_product_numpy = np.dot(array1, array2)

end = time.time()

print("Dot product using NumPy:", dot_product_numpy)
print("Time taken using NumPy:", end - start, "seconds")
```

#4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
import time
import random

n = 1000

A = [[random.random() for _ in range(n)] for _ in range(n)]
B = [[random.random() for _ in range(n)] for _ in range(n)]

C = [[0 for _ in range(n)] for _ in range(n)]

start = time.time()

for i in range(n):
    for j in range(n):
        for k in range(n):
            C[i][j] += A[i][k] * B[k][j]

end = time.time()

print("Time taken using Python lists:", end - start, "seconds")
```

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
import numpy as np
import time

n = 1000

A_np = np.random.rand(n, n)
B_np = np.random.rand(n, n)

start = time.time()

C_np = np.dot(A_np, B_np)

end = time.time()

print("Time taken using NumPy:", end - start, "seconds")
```
